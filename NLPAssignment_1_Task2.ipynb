```python
import re
import pandas as pd
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

**Input Answers**

```python
input_answers = open('STS2016.input.answer-answer.txt','r')
input_headlines = open('STS2016.input.headlines.txt','r')
input_plagiarism = open('STS2016.input.plagiarism.txt','r')
input_postediting = open('STS2016.input.postediting.txt','r')
input_questions = open('STS2016.input.question-question.txt','r')

#Reading content from the text file
content = input_answers.readlines()

#Stripping each line in the content
content = [x.strip() for x in content]

#Removing the URL and unnecessary data from the text
content_split = [i.split('\tStack')[0] for i in content]

#Splitting each line into two sentences by removing tab space
result = [item.split('\t') for item in content_split]
#print(len(result))
```

```python
#Creating a dataframe with the sentence list
import pandas as pd
df=pd.DataFrame(result)
df.columns = ["sentence1","sentence2"]
#df
```

```python
#Removing stopwords from the dataframe columns
stop = stopwords.words('english')
df['s1_cleaned'] = df['sentence1'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
df['s2_cleaned'] = df['sentence2'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
#Printing the dataframe into a list
list_answers=df.values.tolist()
#print(list_answers)

#Creating two lists with the cleaned columns from the dataframe
sentence1=[]
sentence2=[]
for i in list_answers:
  sentence1.append(i[2])
  sentence2.append(i[3])


print(sentence1)
print(sentence2)
```

**Headlines**

```python
#For Headlines
content_headlines = input_headlines.readlines()
content_h = [x.strip() for x in content_headlines]
content_headlines = [i.split('\tEurope Media')[0] for i in content_h]
result_headlines = [item.split('\t') for item in content_headlines]
df_headlines=pd.DataFrame(result_headlines)
df_headlines.columns = ["headline1","headline2"]
stop = stopwords.words('english')
df_headlines['h1_cleaned'] = df_headlines["headline1"].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
df_headlines['h2_cleaned'] = df_headlines['headline2'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
list_headlines=df_headlines.values.tolist()
headline1=[]
headline2=[]
for i in list_headlines:
 headline1.append(i[2])
 headline2.append(i[3])

 #print(list_headlines)
```

**Plagiarism**

```python
#For Plagiarism
content_plagiarism = input_plagiarism.readlines()
content_p = [x.strip() for x in content_plagiarism]
content_plagiarism = [i.split('\tA Corpus')[0] for i in content_p]
result_plagiarism = [item.split('\t') for item in content_plagiarism]
df_plagiarism=pd.DataFrame(result_plagiarism)
df_plagiarism.columns = ["plag1","plag2"]
stop = stopwords.words('english')
df_plagiarism['p1_cleaned'] = df_plagiarism["plag1"].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
df_plagiarism['h2_cleaned'] = df_plagiarism['plag2'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
list_plagiarism=df_plagiarism.values.tolist()
plag1=[]
plag2=[]
for i in list_plagiarism:
 plag1.append(i[2])
 plag2.append(i[3])
```

**Post Editing**

```python
#For PostEditing
content_postediting = input_postediting.readlines()
content_po = [x.strip() for x in content_postediting]
content_postediting = [i.split('\tEAMT11')[0] for i in content_po]
result_postediting = [item.split('\t') for item in content_postediting]
df_postediting=pd.DataFrame(result_postediting)
df_postediting.columns = ["p1","p2"]
stop = stopwords.words('english')
df_postediting['pe1_cleaned'] = df_postediting["p1"].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
df_postediting['pe2_cleaned'] = df_postediting['p2'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
list_postediting=df_postediting.values.tolist()
pe1=[]
pe2=[]
for i in list_postediting:
 pe1.append(i[2])
 pe2.append(i[3])
```

**Question**

```python
#For Question
content_question = input_questions.readlines()
content_q = [x.strip() for x in content_question]
content_question = [i.split('\tStackExchange')[0] for i in content_q]
result_questions = [item.split('\t') for item in content_question]
df_questions=pd.DataFrame(result_questions)
df_questions.columns = ["question1","question2"]
stop = stopwords.words('english')
df_questions['q1_cleaned'] = df_questions['question1'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
df_questions['q2_cleaned'] = df_questions['question2'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
list_questions=df_questions.values.tolist()
question1=[]
question2=[]
for i in list_questions:
 question1.append(i[2])
 question2.append(i[3])
```

**SBERT**

```python
from sentence_transformers import SentenceTransformer
def sbert(sent1, sent2):
  sbert_model = SentenceTransformer('bert-base-nli-mean-tokens')
  embeddings1 = sbert_model.encode(sent1, convert_to_tensor=True)
  embeddings2 = sbert_model.encode(sent2, convert_to_tensor=True)
  return embeddings1, embeddings2
```

```python
s1_embeddings, s2_reshaped = sbert(sentence1, sentence2)
```

```python
h1_embeddings, h2_embeddings = sbert(headline1, headline2)
```

```python
p1_embeddings, p2_embeddings = sbert(plag1, plag2)
```

```python
pe1_embeddings, pe2_embeddings = sbert(pe1, pe2)
```

```python
q1_embeddings, q2_embeddings = sbert(question1, question2)
```

```python
from sentence_transformers import SentenceTransformer, util
def c_score(embed1,embed2):
  cosine_scores = util.cos_sim(embed1, embed2)
  return cosine_scores
```

```python
#Cosine similarity for input.answer-answer
ans_cosine = c_score(s1_embeddings, s2_reshaped)
print(ans_cosine)
ans_cosine_list= ans_cosine.tolist()

#Writing the cosine similarity to a file
ans= open(r'answerCosine.txt', 'w')
for item in ans_cosine_list:
  for i in item:
    ans.write("%s\n" %i)
```

```python
#Cosine similarity for input.headlines
head_cosine = c_score(h1_embeddings, h2_embeddings)
print(head_cosine)
head_cosine_list= head_cosine.tolist()

head= open(r'headlinesCosine.txt', 'w')
for item in head_cosine_list:
  for i in item:
    head.write("%s\n" %i)
```

```python
#Cosine similarity for input.plagiarisgm
plag_cosine = c_score(p1_embeddings, p2_embeddings)
print(plag_cosine)
plag_cosine_list= plag_cosine.tolist()

plag= open(r'plagiarismCosine.txt', 'w')
for item in plag_cosine_list:
  for i in item:
    plag.write("%s\n" %i)
```

```python
#Cosine similarity for input.postediting
pe_cosine = c_score(pe1_embeddings, pe2_embeddings)
print(pe_cosine)
pe_cosine_list= pe_cosine.tolist()

pe= open(r'posteditingCosine.txt', 'w')
for item in pe_cosine_list:
  for i in item:
    pe.write("%s\n" %i)
```
